```python
import pandas as pd
import os
os.chdir(r'C:\Users\cyrus\Case_Study_Coursera\cleaning data')
for f in os.listdir(r'C:\Users\cyrus\Case_Study_Coursera\cleaning data'):
	print(f)
df202004 = pd.read_csv("202004-divvy-tripdata.csv")
df202005 = pd.read_csv("202005-divvy-tripdata.csv")
df202006 = pd.read_csv("202006-divvy-tripdata.csv")
df202007 = pd.read_csv("202007-divvy-tripdata.csv")
df202008 = pd.read_csv("202008-divvy-tripdata.csv")
df202009 = pd.read_csv("202009-divvy-tripdata.csv")
df202010 = pd.read_csv("202010-divvy-tripdata.csv")
df202011 = pd.read_csv("202011-divvy-tripdata.csv")
df202012 = pd.read_csv("202012-divvy-tripdata.csv")
df202101 = pd.read_csv("202101-divvy-tripdata.csv")
df202102 = pd.read_csv("202102-divvy-tripdata.csv")
df202103 = pd.read_csv("202103-divvy-tripdata.csv")
df = [df202004,df202005,df202006,df202007,df202008,df202009,df202010,df202011,df202012,df202101,df202102,df202103]
df = pd.concat(df)
```

```python
print(df.info())
print(df.describe)
```

```python
del df['start_lat+lng']
del df['end_lat+lng']
del df['trip_length']
```

```python
clean_df = df.dropna(axis = 0, how ='any') 
print("Old data frame length:", len(df), "\nNew data frame length:",  
       len(clean_df), "\nNumber of rows with at least 1 NA value:", 
       (len(df)-len(clean_df)))

clean_df.head
```

```python
day = {
    1:"Sunday",
    2:"Monday",
    3:"Tuesday",
    4:"Wednesday",
    5:"Thursday",
    6:"Friday",
    7:"Saturday"
}
clean_df["week_day"].replace(day, inplace=True)
clean_df.head
    
```
